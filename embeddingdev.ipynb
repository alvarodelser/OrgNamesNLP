# LABEL EMBEDDINGS

```python
from orgpackage.clusterembedder import embed_prototypes
rows = []
for domain, labels in DOMAIN_CLASSES_CORR.items():
    for label in labels:
        rows.append({"domain": domain, "label": label.replace("_", " ")})

df = pd.DataFrame(rows)

for model_key in ['fasttext', 'qwen', 'mistral', 'multilingual-e5', 'e5-small']:
    print(f"Embedding with model: {model_key}")
    embeddings = embed_prototypes(df["label"].tolist(), model_key=model_key)
    df[f"{model_key}_embeddings"] = embeddings

df.to_csv("results/embeddings/label_embeddings.csv", index=False)
```
