# Multi-Model Embedding Analysis
This notebook performs a memory-efficient and statistically optimized comparative analysis of all organizational embedding models. 

```python
import numpy as np
import pandas as pd
import os
import gc
import time
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from orgpackage.aux import load_dataset, load_embeddings
from orgpackage.config import DOMAIN_CLASSES_CORR

import warnings
warnings.filterwarnings('ignore')

# Global Plotting Settings
plt.rcParams.update({
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.dpi': 150,
    'savefig.dpi': 300,
    'axes.spines.top': False,
    'axes.spines.right': False
})

os.makedirs("figures", exist_ok=True)

MODELS = ['e5-small', 'multilingual-e5', 'qwen', 'mistral', 'finetuned-me5']

exp_classes = [c for sublist in DOMAIN_CLASSES_CORR.values() for c in sublist]
exp_classes = sorted(list(set(exp_classes)), key=len, reverse=True)
```

## Optimized Metric and Utility Functions
Uses **Query Sampling** ($N=5,000$) and vectorized rank calculations to handle large datasets ($N>60,000$) efficiently.

```python
def compute_metrics(X, y, k=10, n_queries=5000):
    # 1. Filter to Experimental Classes
    mask = y != "Others"
    X_f, y_f = X[mask].astype('float32'), y[mask]
    N_total = len(X_f)
    if N_total == 0: return np.nan, np.nan, np.nan, np.nan, [0.0]*20
    
    # --- Global Stats (FDR & Isotropy) ---
    fdr = np.nan
    unique_classes = np.unique(y_f)
    if len(unique_classes) > 1:
        dim = X.shape[1]
        global_mean = np.mean(X_f, axis=0)
        S_W, S_B = np.zeros((dim, dim)), np.zeros((dim, dim))
        for c in unique_classes:
            Xc = X_f[y_f == c]
            if len(Xc) <= 1: continue
            mc = np.mean(Xc, axis=0)
            S_W += (Xc - mc).T @ (Xc - mc)
            diff = (mc - global_mean).reshape(-1, 1)
            S_B += len(Xc) * (diff @ diff.T)
        fdr = np.trace(np.linalg.pinv(S_W) @ S_B)

    Xc_centered = X_f - np.mean(X_f, axis=0)
    cov = np.cov(Xc_centered, rowvar=False)
    evs = np.linalg.eigvalsh(cov)
    isotropy = evs.min() / evs.max() if evs.max() > 0 else 0

    # --- Sampled IR Metrics (MRR & Recall) ---
    # We sample n_queries to act as 'probes' into the full database
    n_q = min(N_total, n_queries)
    q_idx = np.random.choice(N_total, n_q, replace=False)
    X_q, y_q = X_f[q_idx], y_f[q_idx]

    print(f"   [Metrics] Computing similarity for {n_q} queries against {N_total} candidates...")
    sims = cosine_similarity(X_q, X_f)
    
    # Mask self-comparison
    for i, idx in enumerate(q_idx):
        sims[i, idx] = -np.inf

    mrr_sum = 0
    ranks = []
    for i in range(n_q):
        # Use argpartition for speed if we only needed recall@k, 
        # but for MRR we sort the whole row to find exact rank.
        sorted_indices = np.argsort(sims[i])[::-1]
        matches = (y_f[sorted_indices] == y_q[i])
        if np.any(matches):
            first_hit_rank = np.argmax(matches) + 1
            mrr_sum += 1.0 / first_hit_rank
            ranks.append(first_hit_rank)
        else:
            ranks.append(np.inf)

    ranks = np.array(ranks)
    mrr = mrr_sum / n_q
    r10 = np.mean(ranks <= 10)
    recall_curve = [float(np.mean(ranks <= rk)) for rk in range(1, 21)]

    return float(np.real(fdr)), isotropy, mrr, r10, recall_curve

def get_target_label(row):
    for c in exp_classes:
        if row.get(c) == 1: return c
    return "Others"
```

## Sequential Processing Loop

```python
full_df = load_dataset()
results = {}

for model in MODELS:
    start_time = time.time()
    print(f"\n{'='*50}")
    print(f"PROCESSING MODEL: {model}")
    print(f"{'='*50}")
    
    path = f"./results/embeddings/{model}_embeddings.csv"
    if not os.path.exists(path):
        print(f"[SKIP] File not found: {path}")
        continue

    print(f"[1/4] Loading embeddings file... ({os.path.getsize(path)/1e9:.2f} GB)")
    emb_df = load_embeddings(path)
    embedding_col = [col for col in emb_df.columns if col.endswith('_embedding')][0]
    
    print("[2/4] Merging with ground truth and mapping labels...")
    m_df = full_df.merge(emb_df[['instance', embedding_col]], on='instance', how='inner')
    m_df = m_df[m_df[embedding_col].notna()].reset_index(drop=True)
    
    X = np.stack(m_df[embedding_col].values)
    m_df['target_label'] = m_df.apply(get_target_label, axis=1)
    y = m_df['target_label'].values
    
    print(f"[3/4] Calculating metrics for {len(X)} records...")
    fdr, iso, mrr, r10, recall_history = compute_metrics(X, y, k=10, n_queries=5000)
    
    print("[4/4] Generating PCA projection (3,000 samples)...")
    n_sample = 3000
    idx = np.random.choice(len(X), min(len(X), n_sample), replace=False)
    pca = PCA(n_components=2)
    X_2d = pca.fit_transform(X[idx])
    
    results[model] = {
        'metrics': {'FDR': fdr, 'Isotropy': iso, 'MRR': mrr, 'Recall@10': r10},
        'recalls': recall_history,
        'pca_points': X_2d,
        'pca_labels': y[idx],
        'pca_var': pca.explained_variance_ratio_
    }
    
    print(f"DONE! Model {model} processed in {time.time() - start_time:.1f}s.")
    print("Clearing memory to baseline...")
    del emb_df, m_df, X, y, X_2d
    gc.collect()
```

## 1. Comparative Metrics Table

```python
metrics_df = pd.DataFrame({m: results[m]['metrics'] for m in results}).T
metrics_df.index.name = "Model"
display(metrics_df)
metrics_df.to_csv("results/comparative_embedding_metrics.csv")
```

## 2. Multi-Model Recall Comparison

```python
plt.figure(figsize=(10, 6))
for model in results:
    plt.plot(range(1, 21), results[model]['recalls'], label=model, lw=1.5, marker='.', markersize=4)

plt.xlabel('k')
plt.ylabel('Recall')
plt.xticks(range(1, 21, 2))
plt.ylim(0, 1.05)
plt.legend(frameon=False, loc='lower right')
plt.title("Comparative Recall@k Performance")
plt.tight_layout()
plt.savefig("figures/comparative_recall_all_models.pdf")
plt.show()
```

## 3. PCA Subplot Comparison

```python
n_models = len(results)
fig, axes = plt.subplots(1, n_models, figsize=(5*n_models, 5), sharex=True, sharey=True)
if n_models == 1: axes = [axes]

cmap = plt.get_cmap('tab10')
target_labels = sorted([l for l in exp_classes])

for i, model in enumerate(results):
    ax = axes[i]
    X_2d = results[model]['pca_points']
    y_v = results[model]['pca_labels']
    
    # Plot Others in light grey background
    mask_o = y_v == "Others"
    ax.scatter(X_2d[mask_o, 0], X_2d[mask_o, 1], c='#f0f0f0', s=10, alpha=0.3, zorder=1)
    
    # Plot Experimental Classes in foreground
    for j, lab in enumerate(target_labels):
        mask = y_v == lab
        if np.any(mask):
            ax.scatter(X_2d[mask, 0], X_2d[mask, 1], label=lab if i==0 else "", 
                       s=20, alpha=0.9, color=cmap(j % 10), zorder=2)
    
    ax.set_title(model)
    ax.set_xlabel(f"PC1 ({results[model]['pca_var'][0]:.1%})")
    if i == 0: ax.set_ylabel(f"PC2 ({results[model]['pca_var'][1]:.1%})")

fig.legend(bbox_to_anchor=(1.01, 0.5), loc='center left', frameon=False)
plt.tight_layout()
plt.savefig("figures/comparative_pca_subplots.pdf", bbox_inches='tight')
plt.show()
```
